```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# Load the data
file_path = (r"C:\Users\ASUS\Downloads\Credit Card Defaulter Prediction(in).csv")
data = pd.read_csv(file_path)
```

```python
# Display the column names
print(data.columns)
```

```python
# Correct target column name
target_column = 'default '
```

```python
# Data Preprocessing
# Handle missing values (if any)
data = data.dropna()
```

```python
# Encode categorical variables
le_sex = LabelEncoder()
data['SEX'] = le_sex.fit_transform(data['SEX'])

le_education = LabelEncoder()
data['EDUCATION'] = le_education.fit_transform(data['EDUCATION'])

le_marriage = LabelEncoder()
data['MARRIAGE'] = le_marriage.fit_transform(data['MARRIAGE'])

le_default = LabelEncoder()
data[target_column] = le_default.fit_transform(data[target_column])
```

```python
# Define features and target
X = data.drop(['ID', target_column], axis=1)
y = data[target_column]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model Training
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_prob = model.predict_proba(X_test)[:, 1]
```

```python
# Evaluation using K-S chart
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
ks_statistic = max(tpr - fpr)
roc_auc = roc_auc_score(y_test, y_pred_prob)
```

```python
# Plot K-S Chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(thresholds, tpr, label='True Positive Rate')
plt.plot(thresholds, fpr, label='False Positive Rate')
plt.xlabel('Thresholds')
plt.ylabel('Rate')
plt.title('K-S Chart')
plt.legend()
plt.show()
```

```python
print(f'K-S Statistic: {ks_statistic}')
print(f'ROC AUC: {roc_auc}')
```
